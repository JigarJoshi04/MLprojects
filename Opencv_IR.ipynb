```
import matplotlib.pyplot as plt
import glob
import numpy as np
%cd ../content/images
X_data = []
files = glob.glob ("*.jpg")
file=sorted(files)
for myFile in file:
    #print(myFile)
    image = plt.imread (myFile)
    X_data.append (image)
%cd ..
```

```
import cv2
import numpy as np
from numpy import nan
#import matplotlib.pyplot as plt
# Get user supplied values
#imagePath = '4.jpg'
#cascPath = 'haarcascade_frontalface_default.xml'

# Create the haar cascade
faceCascade = cv2.CascadeClassifier('../usr/local/lib/python3.6/dist-packages/cv2/data/haarcascade_frontalface_default.xml')

counter = 0
result=[]
for image in X_data:
  try:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )
    
    print("Found  faces! {0}".format(len(faces)) + "  at image  " + str(counter))
    result.append(format(len(faces)))
    counter += 1
    
  except cv2.error as e:
    counter += 1
    result.append(nan)
    print("The image "+ str(counter)+ " is not working.")
    
    counter += 1
    

# Draw a rectangle around the faces
#for (x, y, w, h) in faces:
    #cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    #roigray = gray[y:y+h, x:x+w]
    #roicolor = image[y:y+h, x:x+w]

#plt.grid(None)
#plt.xticks([])
#plt.yticks([])
#imgplot = plt.imshow(image)

#cv2.imshow("Faces found", image)
#cv2.waitKey(0)
```
